```python
import pandas as pd

pkgs = pd.read_csv("data/packages.csv.")
pkgsu = pd.read_csv("data/packages_undeployed.csv")
```

```python
pkgs_first_loser = pkgs.loc[(pkgs["first_place"] == True) & (pkgs["winner"] == False)]

pkgs_first_loser_num = pkgs_first_loser.shape[0]
pkgs_num = pkgs.shape[0]
```

```python
pc_pkgs_first_loser = pkgs_first_loser_num / pkgs_num * 100

print('Percentage of packages that were first place but did not make it to the website: ',pc_pkgs_first_loser, "%")
```

```python
#pkgs.head()
```

```python
# check difference of days between "created_at" and "test_week"
    
with_image = pkgs[pkgs["share_image"].notnull()] # packages with image

with_image_num = with_image.shape[0]

with_image_pc = with_image_num / pkgs_num * 100

print('Percentage of packages with images: ',with_image_pc, "%")

#------------------------------------------------------------------

with_image_f_l = with_image.loc[(with_image["first_place"] == True) & (with_image["winner"] == False)] # packages with images that got first place but lost

with_image_f_l_num = with_image_f_l.shape[0]

with_image_f_l_num_pc = with_image_f_l_num / with_image_num * 100

print('Percentage of packages with images that got first place but did not make it to the website: ',with_image_f_l_num_pc, "%")

print('There is minimal improvement in comparison to packages without images: +', with_image_f_l_num_pc-pc_pkgs_first_loser, "%")
```
